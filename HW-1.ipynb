```python
# Problem 2.1
def list_types(input_list):
    l = list()
    for item in input_list:
        l.append(type(item))
    return l
```

```python
# Problem 3
def calc_avg(input_list):
    """ Initialize count and summation variables """
    count = 0
    summation = 0

    """ Loop through input_list to generate a summation and count  """
    for item in input_list:
        # Add the current list item to summation
        summation += item
        # Add one to count
        count += 1
        """account for the possibility of zero as a denominator"""
    if count == 0:
        return 0
        """Calculate the average"""
    else:
        average = summation / count
        return average
```

```python
calc_avg(input_list = [1, 2, 3, 4, 44])
```
